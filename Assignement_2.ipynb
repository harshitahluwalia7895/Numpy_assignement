```python
#Answer_1

import numpy as n
arr = n.random.random((10, 1))
m=n.mean(arr)
print("The array is:\n{}\n The Mean is:{}".format(arr,m))
```

```python
#Answer_2

import numpy as n
array = n.random.random((20, 1))
s=n.std(array)
new=s**2
print("The array is:\n{}\n The Variance is:{}\n The Standard Deviation is:{}".format(array,new,s))
```

```python
#Answer_3

import numpy as n
A = n.random.random((10,20))
print("Matrix A is: \n",A,"\n")
B = n.random.random((20,25))
print("Matrix B is: \n",B,"\n")
C=n.matmul(A, B)
print("Multiplication of a and B: \n",C,"\n")
print("Sum of elements of C:",n.sum(C))
```

```python
#Answer_4
import numpy as n
A = n.random.random((10,1))

print("The original matrix is:\n{}\n".format(A))

def fun(num):
    return (1 / (1 + n.exp(-num)))

m = n.apply_along_axis(fun, 0, A)
print("The result matrix is:\n{}\n".format(m))
```
